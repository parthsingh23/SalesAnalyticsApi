# Calculating KPIs

```python
df["revenue"] = df["price_unit"] * df["units_sold"]
```

<h3>1. Total Revenue</h3>

```python
print(df["revenue"].sum())
```

<h3>2. Total Units Sold</h3>

```python
print(df["units_sold"].sum())
```

<h3>3. Average Selling Price</h3>

```python
print(df["price_unit"].mean())
```

<h3>4. Best Selling Region</h3>

```python
best_region = df.groupby("region")["units_sold"].sum().sort_values(ascending=False)
print(best_region)
```

<h3>5. Best Selling Category</h3>

```python
best_category = df.groupby("category")["units_sold"].sum().sort_values(ascending=False)
print(best_category)
```

<h3>6. Top 10 sku
</h3>


```python
print(df.groupby("sku")["revenue"].sum().sort_values(ascending=False).head(10))
```

<h3>7. Revenue by Channel
</h3>

```python
print(df.groupby("channel")["revenue"].sum())
```

# List Comprehensions

```python
upper = [col.upper() for col in df.columns]
print(upper)
```

```python
numeric_cols = [col for col in df.columns if (df[col]).dtype not in ['str', 'object', 'string']]
print(numeric_cols)
```
